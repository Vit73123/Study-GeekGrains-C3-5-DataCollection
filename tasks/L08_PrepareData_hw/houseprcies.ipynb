### Пропущенные значения

Удалим строки с единичными пропущенными данными:

```python
dfa = df.dropna(subset=['Utilities',
    'Exterior1st',
    'Exterior2nd',
    'BsmtFinSF1',
    'BsmtFinSF2',
    'BsmtUnfSF',
    'TotalBsmtSF',
    'Electrical',
    'BsmtFullBath',
    'BsmtHalfBath',
    'KitchenQual',
    'Functional',
    'GarageCars',
    'GarageArea',
    'SaleType',
    'SaleCondition'])
```

В столбце `LotFrontage` с числовым некотегориальным признаком пропущенные данные заменим медианой:

```python
# Подавить предупреждение SettingWithCopyWarning:
pd.options.mode.chained_assignment = None

median_lotfrontage = dfa['LotFrontage'].median()
dfa['LotFrontage'] = dfa['LotFrontage'].fillna(median_lotfrontage)
```

```python
dfa['LotFrontage'].info()
```

В остальных столбцах с пропущеными данными, данные оставляем как Null, т.к. фактически эти данные не являются пропущщенными, т.е. имеют смысловую нагрузку - предполагают отсутствие связанного с ними признака (в соответствии с прилагаемым описанием данных).

## Разведочный анализ (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

#### Распределение SalePrice

```python
dfa['SalePrice'].describe()
```

Цена существенно отклоняется от нормального распределения:
- слишком большое стандартное отклонение;
- минимум больше 0, т.к. в ряду только положительные числа (цена);
- слишком большая разница между минимальным значением и 25-м процентилем;
- разница между 75-м процентилем и максимумом больше, чем 25-й процентиль и максимум.

Сведения о цене нельзя считать выбросами, т.к. каждая цена является значимым показателем.  
При этом отдельные значения отклоняются значительно.

```python
sns.catplot(data=dfa['SalePrice'])
```

Построим гистограмму для визуального отображения распределения:

```python
sns.histplot(dfa['SalePrice'], kde=True)
```

Асимметрия и эксцесс:

```python
print("Ассиметрия: %f" % dfa['SalePrice'].skew())
print("Эксцесс: %f" % dfa['SalePrice'].kurt())
```

Проверим как повлияет логарифмическое преобразование целевой переменной для уменьшения асимметрии:

```python
fig = plt.figure(figsize = (10,5))

# Распределение на необработанных данных
fig.add_subplot(1,2,1)
res = stats.probplot(dfa['SalePrice'], plot=plt)

# Распределение при условии, что мы прологарифмировали 'SalePrice'
fig.add_subplot(1,2,2)
res = stats.probplot(np.log1p(dfa['SalePrice']), plot=plt)
```

Логарифмическое преобразование цены исправило асимметрию.  
Применим данное преобразование к датасету:

```python
dfa['SalePrice'] = np.log1p(dfa['SalePrice'])
```

```python
sns.histplot(dfa['SalePrice'], kde=True)
```

#### Корреляция

Матрица корреляции для 10 коррелирующих признаков:

```python
k = 5 # количество коррелирующих признаков, которое мы хотим увидеть
corr_mat = dfa.corr(numeric_only=True)
cols = corr_mat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(dfa[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, 
                 fmt='.2f', annot_kws={'size': 10}, 
                 yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

У `SalePrice` наименьшая корреляция с признаками `GrLiveArea`, `OverallQual`.  
С `GarageCars` цена коррелирует значительно, поэтому изменение этих данных может значительно повлиять на результат анализа `SalePrice`, т.е. преобразования этими данными производить нельзя.

#### Выбросы

## Проектирование признаков

```python
from sklearn.preprocessing import OrdinalEncoder
import datetime
```

#### One-hot encoding

Список полей категориальных признаков, к которым будет применим one-hot encoding:

```python
oh_cols = ['Street',
           'Alley',
           'LotShape',
           'LandContour',
           'Utilities',
           'LotConfig',
           'LandSlope',
           'LandSlope',
           'CentralAir'
          ]
```

```python
dfa = pd.get_dummies(dfa, columns=oh_cols)
```

Список полей категориальных признаков, к которым будет применим label encoding:

```python
le_cols = ['MSSubClass',
           'MSZoning',
           'Neighborhood',
           'Condition1',
           'Condition2',
           'BldgType',
           'HouseStyle',
           'OverallQual',
           'OverallCond',
           'RoofStyle',
           'RoofMatl',
           'Exterior1st',
           'Exterior2nd',
           'MasVnrType',
           'ExterQual',
           'ExterCond',
           'Foundation',
           'BsmtQual',
           'BsmtCond',
           'BsmtExposure',
           'BsmtFinType1',
           'BsmtFinType2',
           'Heating',
           'HeatingQC',
           'Electrical',
           'KitchenQual',
           'Functional',
           'FireplaceQu',
           'GarageType',
           'GarageFinish',
           'GarageQual',
           'GarageCond',
           'PavedDrive',
           'PoolQC',
           'Fence',
           'MiscFeature',
           'SaleType',
           'SaleCondition',
          ]
```

```python
oe = OrdinalEncoder()
dfa[le_cols] = oe.fit_transform(dfa[le_cols])
```
